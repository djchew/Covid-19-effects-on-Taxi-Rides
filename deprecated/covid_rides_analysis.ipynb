```python
months = ["March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
```

```python
covid_files = ["march_covid.csv","april_covid.csv","may_covid.csv","june_covid.csv","july_covid.csv","august_covid.csv",
         "september_covid.csv", "october_covid.csv", 
         "november_covid.csv","december_covid.csv"]
covid_csv = pd.concat([pd.read_csv(file) for file in covid_files])
```

```python
covid_csv = covid_csv.replace("StatenIsland", "Staten Island")
```

```python
covid_csv.rename(columns = ({"BOROUGH_GROUP": "borough"}), inplace= True)
```

```python
boroughs_covid_count = ["Bronx_Count", "Brooklyn_Count", "Queens_Count", "Manhattan_Count", "StatenIsland_Count"]
```

```python
bronx_covid_csv = covid_csv.loc[covid_csv["borough"] == "Bronx"]
brooklyn_covid_csv = covid_csv.loc[covid_csv["borough"] == "Brooklyn"]
queens_covid_csv = covid_csv.loc[covid_csv["borough"] == "Queens"]
manhattan_covid_csv = covid_csv.loc[covid_csv["borough"] == "Manhattan"]
statenisland_covid_csv = covid_csv.loc[covid_csv["borough"] == "Staten Island"]
```

```python
borough_covid_list = [bronx_covid_csv, brooklyn_covid_csv, queens_covid_csv, manhattan_covid_csv , statenisland_covid_csv]
for i in range(len(borough_covid_list)):
    borough_covid_list[i].insert(3,"Month", months)
    borough_covid_list[i].rename(columns = {"CASE_RATE": boroughs_covid_count[i]}, inplace = True)
```

```python
covid_df = pd.DataFrame(data = {"Months": months, "Bronx": bronx_covid_csv["Bronx_Count"].tolist(),
                          "Queens": queens_covid_csv["Queens_Count"].tolist(),
                  "Brooklyn": brooklyn_covid_csv["Brooklyn_Count"].tolist(),
                          "Manhattan": manhattan_covid_csv["Manhattan_Count"].tolist(),
                  "Staten Island": statenisland_covid_csv["StatenIsland_Count"].tolist()})
```

```python
plt.clf()
```
